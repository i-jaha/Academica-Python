```python
import numpy as np
```

```python
# Создай одномерный массив из 5 целых чисел.
array_1 = np.random.randint(10, size=5)
# взял отсюда https://numpy.org/doc/2.1/reference/random/generated/numpy.random.randint.html
print(array_1)
```

```python
# Создай трехмерный массив размером 3x3х3, то есть должно быть три матрицы с размером  3х3.
array_2 = np.random.randint(1, 100, size=(3, 3, 3))
print(array_2)
```

```python
# Создай одномерный массив из 8 случайных целых чисел в диапазоне от 1 до 20.
array_3 = np.random.randint(1, 20, size=8)
print(array_3)
```

```python
# Найди и выведи второй элемент вашего одномерного массива.
print(f"Второй элемент одномерного массива \n {array_3[1]}\n")
```

```python
# Найди и выведи значение во второй строке и третьем столбце первой матрицы вашего трехмерного массива.
print(f"Первая матрица трехмерного массива \n {array_2[0]}\n")
print(f"Зачение 2ой строки и 3его столбца 1ой матрицы трехмерного массива \n {array_2[0, 1, 2]}\n")
```

```python
# Поменяйте любое значение 2 матрицы трехмерного массива
print(f"До изменения \n {array_2}\n")
array_2[:, 1, :] = 0
# поменял на 0 (нуль) все значения во вторых строках всех матриц
print(f"После изменения \n {array_2}\n")
```

```python
# Создай одномерный массив от 1 до 9. 
array_4 = np.arange(1, 10)
# Используя срезы, выведи элементы от 3 до 9
print(array_4[2:9])
```

```python
array_5 = np.random.randint(1, 100, size=(5, 4))
print(f"Двумерная матрица создана: \n{array_5}\n")
print(f"Первые три строки матрицы: \n{array_5[:3, :]}\n")
print(f"Последний столбец матрицы: \n{array_5[:, -1]}\n")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

picture = cv2.imread("cat.jpg")
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)

edited_picture = np.copy(picture)
edited_picture[:, :, 0] = 70 # Red
edited_picture[:, :, 1] = 40 # Green
bright = edited_picture[:, :, 2] > 120
edited_picture[bright, 2]= 255

plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.imshow(picture)
plt.title("Original image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edited_picture)
plt.title("Edited image")
plt.axis("off")

plt.show()
```
