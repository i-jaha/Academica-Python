```python
import pandas as pd
```

```python
df = pd.read_csv("ДЗ_Pandas.csv")
```

```python
# Определите, в каких столбцах есть пропущенные значения. 
df.isnull().sum()
```

```python
# Удалите все строки, в которых пропущено значение в столбце 'Бюджет (млн $)'.
df.dropna(subset=["Бюджет (млн $)"], inplace = True)
df
```

```python
# Заполните все пропущенные значения в столбце 'Режиссер' строкой 'Неизвестен'.
df["Режиссер"].fillna("Неизвестен", inplace = True)
df
```

```python
# Заполните все пропущенные значения в столбце 'Рейтинг' средним значением этого столбца.
df["Рейтинг"].fillna(df["Рейтинг"].mean(), inplace = True)
df
```

```python
# Если в строке пропущены значения и в 'Бюджет (млн $)', и в 'Сборы (млн $)', удалите эту строку. 
# В противном случае, если пропущено значение только в одном из этих столбцов, оставьте строку.
df.dropna(subset=["Бюджет (млн $)", "Сборы (млн $)"], how = "all", inplace = True)
df
```

```python
# Сгруппируйте фильмы по жанру
grouped_df = df.groupby("Жанр")
```

```python
# Рассчитайте средний рейтинг и средний бюджет в каждом жанре.
print(grouped_df[["Рейтинг", "Бюджет (млн $)"]].mean())
```

```python
# Сгруппируйте фильмы по году выпуска
grouped_df_2 = df.groupby("Год выпуска")
```

```python
# Определите общее количество фильмов и суммарные сборы в прокате по каждому году.
grouped_df_2.agg({"Название фильма": "count", "Сборы (млн $)": "sum"})
```

```python
# Сгруппируйте фильмы по жанру и году выпуска
grouped_df_3 = df.groupby(["Жанр", "Год выпуска"])
```

```python
# Рассчитайте средний рейтинг для каждой комбинации жанра и года выпуска
grouped_df_3["Рейтинг"].mean()
```

```python
# Сгруппируйте фильмы по жанру и рассчитайте средний рейтинг для каждого жанра
grouped_df = df.groupby("Жанр")
```

```python
# рассчитайте средний рейтинг для каждого жанра
grouped_df["Рейтинг"].mean().reset_index()
```

```python
# Сбросьте индекс и отсортируйте результаты по среднему рейтингу в убывающем порядке
grouped_df["Рейтинг"].mean().reset_index().sort_values(by = "Рейтинг", ascending = False)
```

```python
# Используйте value_counts для определения, сколько фильмов было выпущено в каждом году
df["Год выпуска"].value_counts()
```

```python
# Отсортируйте результаты по годам в порядке возрастания
df["Год выпуска"].value_counts().reset_index().sort_values(by = "Год выпуска")
```

```python
# Сгруппируйте фильмы по режиссеру и жанру, затем рассчитайте общие сборы фильмов для каждой комбинации
df.groupby(["Режиссер", "Жанр"])["Сборы (млн $)"].sum().reset_index()
```

```python
# Сбросьте индекс и отсортируйте результаты по сумме сборов в убывающем порядке
df.groupby(["Режиссер", "Жанр"])["Сборы (млн $)"].sum().reset_index().sort_values(by="Сборы (млн $)", ascending=False)
```
