```python
x = 2
for i in range(1, 101):
    x = x + i
    print(f"{i} + {x - i} = {x}")
```

```python
number = int(input("Введите число для таблицы умножения: "))

for i in range(1, 11):
    result = number * i
    print(f"{number} x {i} = {result}")
```

```python
result = []
for i in range(2, 21):
    if i % 2 == 0:
        result.append(i)
print(f"Четные числа: {result}")
```
