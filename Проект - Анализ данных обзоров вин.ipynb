```python
# Удалите колонку 'LikePresentations', она нам не понадобится
df.drop(columns = ["LikePresentation"], inplace = True)
df
```

```python
# Измените названия ваших колонок - разделите слова в названии колонок. 
# Например, замените AcademicPerformance на Academic Performance
column_list = list(df.columns) # получил название колонок
new_columns = []
for name in column_list:
    name = name.replace(" ", "")
    new_name = ""
    for i in name:
        if i.isupper() and new_name:
            new_name = new_name + " " + i
        else: 
            new_name = new_name + i
    new_columns.append(new_name)

df.columns = new_columns
df
```

```python
# Очистка данных. Удалите дубликаты, если они есть ...
df.drop_duplicates(inplace = True)
df
```

```python
# … и строки с пропущенными значениями
df.dropna(how = "any", inplace = True)
df
```

```python
# Выявить среднее количество сна студентов
df["Sleep Per Day Hours"].mean()
```

```python
# Вывести таблицу, где только студенты с оценками "excellent" …
df[df["Academic Performance"] == "Excellent"]
```

```python
# … и посчитать среднее значение сна
df[df["Academic Performance"] == "Excellent"]["Sleep Per Day Hours"].mean()
```

```python
# Вывести таблицу, где только студенты с депрессией - yes …
df[df["Depression Status"] == "Yes"]
```

```python
# … и посчитать среднее значение их сна
df[df["Depression Status"] == "Yes"]["Sleep Per Day Hours"].mean()
```
