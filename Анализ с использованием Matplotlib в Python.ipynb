```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("coffee_sales_data.csv")
df.info()
```

```python
df.head()
```

# Задание 1
Постройте столбчатую диаграмму, показывающую распределение общего количества продаж кофе по дням недели. 

Используйте дни недели на оси X и общее количество продаж на оси Y. 

Какие дни недели имеют наибольшие продажи?

```python
sales_by_weekday = df.groupby("День недели")["Количество Продаж"].sum()
sales_by_weekday
```

```python
sales_by_weekday.plot(kind="bar", color="skyblue")
plt.title("Distribution of coffee sales by day of the week", fontsize=16)
plt.xlabel("Day of the week", fontsize=12)
plt.ylabel("Quantity of sales", fontsize=12)
plt.grid(axis="y")
plt.xticks(rotation=45)
plt.show()
```

### Наибольшие продажи имеют Суббота, Воскресенье и Понедельник

# Задание 2
Создайте boxplot, который покажет распределение цен на разные типы кофе (Эспрессо, Латте, Капучино). 

Используйте тип кофе на оси X и цены на оси Y. 

Какой тип кофе имеет наибольший разброс цен?

```python
df["Тип Кофе"].unique()
```

```python
df.boxplot(column="Цена за Чашку", by="Тип Кофе")

plt.title("Распределение цен на разные типы кофе")
plt.xlabel("Тип кофе")
plt.ylabel("Цена за Чашку")
plt.suptitle('')

plt.show()
```

# Задание 3: Анализ взаимосвязи между выходными и продажами

Исследовать влияние выходных дней на продажи кофе в магазине с помощью столбчатой диаграммы, показывающей разницу в общем количестве продаж между выходными и будними днями. 

Определить, в какие дни наблюдается больше продаж

```python
sales_by_day = df.groupby('Выходной')['Количество Продаж'].sum()

sales_by_day.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Продажи кофе в будние и выходные дни")
plt.xlabel("День")
plt.ylabel("Количество продаж")
plt.xticks([0, 1], ["Будние", "Выходные"], rotation=0)
plt.show()
```
