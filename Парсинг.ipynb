```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "https://tengrinews.kz/"
response = requests.get(link)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
sections = soup.find_all("section")
for section in sections:
    # print("--- Новая секция ---")

    headings = section.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    if headings: 
        # print("Заголовки в секции:")
        for heading in headings:
            print(f"{heading.text.strip()}")
    # else:
        # print("\tЗаголовки в этой секции не найдены.") # Сообщение, если заголовков нет
            
    elements_with_both = section.select('[class*="title"][class*="main-news"]')
    if elements_with_both:
        # print("\tЭлементы с 'title' и 'main-news' в секции:")
        for element in elements_with_both:
            print(f"\t{element.text.strip()}")
    # else:
        # print("\tЭлементы с 'title' и 'main-news' в этой секции не найдены.")
```
