```python
num_students = 1000
grades = np.random.uniform(2, 5, num_students) # сделал не с помощью randint, чтобы получить более широкое распределение
num_bins = np.random.randint(2,8) # Количество бинов

plt.hist(grades, bins=num_bins, edgecolor="black")

plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.title("Grade distribution")

plt.show()
```

```python
coffee_types = ["Espresso", "Latte", "Cappuccino"]
coffee_sales = [50, 80, 120]

snacks = ["Croissants", "Cheesecake", "Muffin"]
snack_sales = [30, 15, 20]

plt.figure(figsize=(15, 5))

ax1 = plt.subplot(1, 2, 1)
ax1.bar(coffee_types, coffee_sales)
ax1.set_xlabel("Coffee")
ax1.set_ylabel("Quantity")
ax1.set_title("Coffee Sales")

ax2 = plt.subplot(1, 2, 2)
ax2.bar(snacks, snack_sales)
ax2.set_xlabel("Snacks")
ax2.set_ylabel("Quantity")
ax2.set_title("Snack Sales")

plt.show()
```
