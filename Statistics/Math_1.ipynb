```python
#1. Создайте функцию через def f(x), где на выходе вы получите x^4 + 3 * x^2 - 15
```

```python
def f(x):
  return x**4 + 3 * x**2 - 15

print(f(2))
```

```python
#2. Создайте список x_points, в котором у нас лежат числа от -5 до 5
```

```python
x_points = list(range(-5, 6))
print(x_points)
```

```python
#3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения
# x_points, также сохраните результаты в y_points
```

```python
y_points = []
for x in x_points:
  y = f(x)
  print(y)
  y_points.append(y)

print(y_points)
```

```python
#4. Покажите на графике x_points & y_points
```

```python
import matplotlib.pyplot as plt

plt.plot(x_points, y_points)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = x^4 + 3x^2 - 15')
plt.grid(True)
plt.show()
```

```python
#5. Импортируйте функцию sin из библиотеки math (from math import sin), далее создайте список x_points, в котором лежат числа
# от -10 до 10. Аналогично создайте y_points используя функцию sin. Также отобразите на графике x_points & y_points
```

```python
import matplotlib.pyplot as plt
from math import sin

x_points = list(range(-10, 11))
y_points = []
for x in x_points:
  y = sin(x)
  y_points.append(y)

plt.plot(x_points, y_points)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('График функции sin(x)')
plt.grid(True)
plt.show()
```
