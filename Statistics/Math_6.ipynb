```python
import numpy as np
import matplotlib.pyplot as plt
```

**Упражнение 1:** Рассмотрим систему уравнений, приведенную ниже. Найдите такое значение коэффициента $a$, чтобы система имела ровно одно решение (то есть найти любое подходящее значение $a$). Для этого значения $a$ найдите решение системы уравнений и начертите линии, образованные вашими уравнениями, чтобы проверить свой ответ.

$$
\begin{eqnarray*}
ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
'''
Для того чтобы система имела ровно одно решение, она должна быть определенной.
Это происходит, когда коэффициенты уравнений не являются пропорциональными.
Чтобы найти значение a, при котором система будет иметь одно решение,
коэффициенты при x1 и x2 не должны быть пропорциональны.
В данном случае a!=1.
Допустим, что a=3.
'''

# Коэффициенты матрицы
A = np.array([[3, 2],
              [4, 8]])

# Вектор правой части
B = np.array([11, 12])

# Решение системы уравнений
solution = np.linalg.solve(A, B)
print("Решение системы уравнений (x1, x2):", solution)

# Создание массива координат x
x = np.linspace(-7, 7, 400)

# Вычисление y для каждого уравнения
y1 = (11 - 3 * x) / 2
y2 = (12 - 4* x) / 8

# Построение графиков
plt.plot(x, y1, label='3x1 + 2x2 = 11')
plt.plot(x, y2, label='4x1 + 8x2 = 12')

# Добавление заголовков и меток осей
plt.xlabel('x')
plt.ylabel('y')
plt.title('График системы уравнений')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(color = 'gray', linestyle = '--')
plt.legend()
plt.show()
```

**Упражнение 2:** Найдите значение коэффициента $a$, при котором система не имеет решений (то есть чтобы функции были параллельны). Постройте линии, образованные вашими уравнениями, чтобы убедиться, что решения нет.

$$
\begin{eqnarray*}
 ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
'''
Для того чтобы система не имела решений, линии, образованные этими уравнениями, должны быть параллельны
Это происходит, когда коэффициенты при x1 и x2 в обоих уравнениях должны быть пропорциональны, но правые части уравнений должны быть различны.
В данном случае a=1.
'''

# Создание массива координат x
x = np.linspace(-10, 10, 400)

# Вычисление y для каждого уравнения при a = 1
y1 = (11 - x) / 2
y2 = (12 - 4 * x) / 8

# Построение графиков
plt.plot(x, y1, label='x1 + 2x2 = 11')
plt.plot(x, y2, label='4x1 + 8x2 = 12')

# Добавление заголовков и меток осей
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('График системы уравнений (параллельные линии)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

**Упражнение 3:** Вычислите $-2E$, $G+F$, $4F-G$, $HG$ и $GE$, используя следующие определения матриц. Сначала выполните упражнение на бумаге, а затем проверьте, выполнив расчет с массивами NumPy. Для умножения матриц используйте np.dot

$$
\begin{equation}
E = \left[ \begin{array}{r} 5 \\ -2 \end{array}\right] \hspace{1cm}
F = \left[ \begin{array}{rr} 1 & 6 \\ 2 & 0 \\ -1 & -1 \end{array}\right] \hspace{1cm}
G = \left[ \begin{array}{rr} 2 & 0\\ -1 & 3 \\ -1 & 6 \end{array}\right] \hspace{1cm}
H = \left[ \begin{array}{rrr} 3 & 0 & 1 \\ 1 & -2 & 2 \\ 3 & 4 & -1\end{array}\right]
\end{equation}
$$

```python
# Определение матриц
E = np.array([[5], [-2]])
F = np.array([[1, 6], [2, 0], [-1, -1]])
G = np.array([[2, 0], [-1, 3], [-1, 6]])
H = np.array([[3, 0, 1], [1, -2, 2], [3, 4, -1]])

# −2E
neg_2E = -2 * E
print("−2E:\n", neg_2E, '\n')

# G + F
G_plus_F = G + F
print("G + F:\n", G_plus_F, '\n')

# 4F - G
four_F_minus_G = 4 * F - G
print("4F − G:\n", four_F_minus_G, '\n')

# H * G
H_dot_G = np.dot(H, G)
print("H * G:\n", H_dot_G, '\n')

# G * E
G_dot_E = np.dot(G, E)
print("G * E:\n", G_dot_E)
```
