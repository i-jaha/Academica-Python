```python
import numpy as np
import matplotlib.pyplot as plt
```

## Задание 1

**Исследование нормальных распределений**
- Сгенерируйте первый список из 1000 случайных чисел, следуя стандартному нормальному распределению (μ=0, σ=1).
    
    Сгенерируйте второй список из 1000 случайных чисел, следуя нормальному распределению с средним значением μ=8 и стандартным отклонением σ=2.
    
- Постройте гистограммы для обоих списков.
    
    Гистограммы должны быть нормализованы, чтобы отражать плотность вероятности.
    
- Оцентрируйте и измените масштаб нормального распределения с μ=8 и σ=2 так, чтобы его среднее значение стало 0, а стандартное отклонение 1.
    
    Постройте гистограмму для модифицированных данных, сравнивая её с гистограммой стандартного нормального распределения.

```python
np.random.seed(42)

mean1 = 0
std1 = 1
sample_size = 1000
list1 = np.random.normal(loc=mean1, scale=std1, size=sample_size)

mean2 = 8
std2= 2
list2 = np.random.normal(loc=mean2, scale=std2, size=sample_size)

print("Первый список (первые 10 значений):")
print(list1[:10])
print("\nВторой список (первые 10 значений):")
print(list2[:10])
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.hist(list1, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title('Гистограмма для первого списка (μ=0, σ=1)')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

plt.subplot(1, 2, 2)
plt.hist(list2, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title('Гистограмма для второго списка (μ=8, σ=2)')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

plt.tight_layout()
plt.show()
```

```python
mean_list2 = np.mean(list2)
std_list2 = np.std(list2)
list2_centered_scaled = (list2 - mean_list2) / std_list2

new_mean = np.mean(list2_centered_scaled)
new_std = np.std(list2_centered_scaled)

print(f'Новое среднее значение: {new_mean:.2f}')
print(f'Новое стандартное отклонение: {new_std:.2f}')
```

```python
plt.figure(figsize=(10, 8))

plt.hist(list1, bins=30, edgecolor='black', alpha=0.7, density=True, label='Стандартное нормальное распределение')
plt.hist(list2_centered_scaled, bins=30, edgecolor='black', alpha=0.7, density=True, label='Модифицированное нормальное распределение')
plt.hist(list2_centered_scaled, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title('Гистограмма для стандартного нормального распределения (μ=0, σ=1)')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

plt.legend()
plt.tight_layout()
plt.show()
```

## Задание 2

**Изучение распределения роста учащихся средней школы и его стандартизация для анализа.**

- Используйте Numpy для создания выборки из 1000 случайных значений роста учащихся, следующих нормальному распределению с средним значением 170 см и стандартным отклонением 10 см.
- Преобразуйте исходные данные так, чтобы получить распределение с средним значением 0 и стандартным отклонением 1, используя простую стандартизацию.
- Постройте гистограммы для исходных данных роста учащихся и для стандартизированных данных.

```python
np.random.seed(42)

mean_height = 170
std_height = 10
sample_size = 1000

heights = np.random.normal(loc=mean_height, scale=std_height, size=sample_size)
print(heights[:10])
```

```python
mean_height2 = np.mean(heights)
std_height2 = np.std(heights)
heights_standardized = (heights - mean_height2) / std_height2

new_mean = np.mean(heights_standardized)
new_std = np.std(heights_standardized)

print(f'Новое среднее значение: {new_mean:.2f}')
print(f'Новое стандартное отклонение: {new_std:.2f}')
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(heights, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title('Гистограмма для исходных данных роста учащихся')
plt.xlabel('Рост (см)')
plt.ylabel('Плотность вероятности')

plt.subplot(1, 2, 2)
plt.hist(heights_standardized, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.title('Гистограмма для стандартизированных данных роста учащихся')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

plt.tight_layout()
plt.show()
```

## Задание 3

**Изменение климата приводит к тому, что погодные условия становятся более переменчивыми, что включает изменения в среднесуточных температурах.**

В этом задании вы будете работать с моделью данных среднесуточных температур для определенного месяца, например, июля. Исходя из исторических данных, предположим, что средняя температура в июле составляет 22°C со стандартным отклонением 3°C.

- Сгенерируйте выборку из 1000 случайных значений среднесуточных температур для июля, используя нормальное распределение с заданными параметрами (среднее = 22°C, стандартное отклонение = 3°C).
- Определите, какой процент дней характеризуется аномально высокими (выше 27°C) и аномально низкими (ниже 15°C) температурами.
- Постройте гистограмму среднесуточных температур для июля.
    
    Выделите области аномальных температур на гистограмме разными цветами.

```python
np.random.seed(42)

mean_temp = 22
std_temp = 3
sample_size = 1000

july_temperatures = np.random.normal(loc=mean_temp, scale=std_temp, size=sample_size)
print(july_temperatures[:10])
```

```python
high_temp = 27
low_temp = 15

high_temp_days = np.sum(july_temperatures > high_temp) / sample_size * 100
low_temp_days = np.sum(july_temperatures < low_temp) / sample_size * 100

print(f'Процент дней с аномально высокими температурами (выше {high_temp}°C): {high_temp_days:.2f}%')
print(f'Процент дней с аномально низкими температурами (ниже {low_temp}°C): {low_temp_days:.2f}%')
```

```python
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(july_temperatures, bins=30, edgecolor='black', alpha=0.7, density=True)

colors = []
for bin in bins[:-1]:
  if bin > high_temp:
    colors.append('lightcoral')
  elif bin < low_temp:
    colors.append('lightblue')
  else:
    colors.append('gray')

for patch, color in zip(patches, colors):
  patch.set_facecolor(color)

plt.axvline(high_temp, color='red', linestyle='--', linewidth=2, label=f'Аномально высокие ({high_temp}°C)')
plt.axvline(low_temp, color='blue', linestyle='--', linewidth=2, label=f'Аномально низкие ({low_temp}°C)')

plt.title('Гистограмма среднесуточных температур для июля')
plt.xlabel('Температура (°C)')
plt.ylabel('Плотность вероятности')

plt.legend()
plt.show()
```
