```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

1. Создайте функцию f(x) = 3 * x^2 + 2 * x - 10

```python
def f(x):
  return 3 * x**2 + 2 * x - 10
```

2. Создайте список x_points (или массив через numpy), в котором у нас лежат числа от -10 до 10

```python
x_points = np.arange(-10, 11)
```

3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения x_points, также сохраните результаты в y_points (cписок или массив)

```python
y_points = []
for x in x_points:
  y = f(x)
  print(f"f({x}) = {y}")
  y_points.append(y)
```

4. Покажите на графике x_points & y_points

```python
plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points, label='f(x)', color='blue')

plt.scatter(x_points, y_points, color='red', zorder=5, label='Точки')

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("График функции f(x) = 3x^2 + 2x - 10")
plt.legend()
plt.grid(True)
plt.show()
```

5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)

```python
x = sp.symbols('x')
f_expr = 3 * x**2 + 2 * x - 10
f = sp.lambdify(x, f_expr, 'numpy')
f_diff = sp.diff(f_expr, x)
print("Производная функции f(x):", f_diff)
```

```python
def f_diff(x):
  return 6*x + 2
```

6. Чему равна производная при x = 5

```python
x_val = 5
y_val = f_diff(x_val)
print(f"Производная f_diff(x) при x = {x_val}: {y_val}")
```

7. Покажите на графике f(x) & f_diff(x)

```python
y_diff_points = []
for x in x_points:
    y_diff_points.append(f_diff(x))

plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points, label='f(x)', color='blue')  # График функции f(x)
plt.plot(x_points, y_diff_points, label="f'(x)", color='red', linestyle='--')  # График производной f'(x)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Графики функций f(x) и f_diff(x)")
plt.legend()
plt.grid(True)
plt.show()
```

8. Найдите уравнение касательной линии в точке x = 8

```python
x_val = 8
y_val = f(x_val)
slope = f_diff(x_val)

# Уравнение касательной: y - y1 = m(x - x1)
# y - y_val = slope * (x - x_val)
# y = slope * (x - x_val) + y_val

print(f"Уравнение касательной в точке x = {x_val}: y = {slope} * (x - {x_val}) + {y_val}")
```

9. Покажите на графике функцию f(x) & касательную линию в точке x = 8

```python
x_tangent = np.linspace(x_val - 2, x_val + 2, 100)
y_tangent = slope * (x_tangent - x_val) + y_val

plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points, label='f(x)', color='blue')
plt.plot(x_tangent, y_tangent, label='Касательная', color='red', linestyle='--')

plt.scatter(x_val, y_val, color='red', zorder=5, label='Точка касания')

plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции f(x) и касательной в точке x = 8")
plt.legend()
plt.grid(True)
plt.show()
```

10. Найдите мин или макс функции f(x) и покажите на графике касательную линию, которая проходит через этот мин или макс функции

```python
min_x = -f_diff(0) / 6
min_y = f(min_x)
slope_at_min = f_diff(min_x)

print(f"Минимум функции в точке x = {min_x}, f(x) = {min_y}")
print(f"Наклон касательной в точке минимума: {slope_at_min}")
print(f"Функция f(x) является квадратичной функцией, и её график представляет собой параболу. Есть только минимум, но нет максимума.")

x_tangent = np.linspace(min_x - 2, min_x + 2, 100)
y_tangent = slope_at_min * (x_tangent - min_x) + min_y

plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points, label='f(x)', color='blue')
plt.plot(x_tangent, y_tangent, label='Касательная', color='red', linestyle='--')
plt.scatter(min_x, min_y, color='red', zorder=5, label='Точка минимума')

plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции f(x) и касательной в точке минимума")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.show()
```

11. Посчитайте интеграл функции f(x) между 0 и 5. Также отобразите на графике площадь, которую мы посчитали.

```python
x = sp.symbols('x')
f_expr = 3 * x**2 + 2 * x - 10
f = sp.lambdify(x, f_expr, 'numpy')
integral = sp.integrate(f_expr, (x, 0, 5))
print(f"Интеграл функции f(x) от 0 до 5: {integral}")

x_vals = np.linspace(0, 5, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')

plt.fill_between(x_vals, y_vals, where=(y_vals >= 0), color='skyblue', alpha=0.5, label='Положительная площадь')
plt.fill_between(x_vals, y_vals, where=(y_vals < 0), color='lightcoral', alpha=0.5, label='Отрицательная площадь')

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(f"Интеграл f(x) от 0 до 5 = {integral}")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.show()
```
