```python
import numpy as np
from scipy.stats import norm
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## Задача 1

**Вероятность распределения**

Рекламное агентство заявляет, что каждая печатная брошюра содержит ровно 20 страниц. Однако, из прошлых данных о качестве печати известно, что количество страниц в брошюре подчиняется нормальному распределению со средним значением 20 страниц и стандартным отклонением 2 страницы. Менеджер по контролю качества хочет проверить, какова вероятность того, что случайно выбранная брошюра будет содержать менее 18 страниц, что будет считаться недопустимым для стандартов агентства.

```python
mean = 20
std_dev = 2
threshold = 18

value = (threshold - mean) / std_dev

# Вычисление вероятности с помощью метода cdf
probability = norm.cdf(value)

print(f'Вероятность того, что брошюра будет содержать менее 18 страниц: {probability*100:.2f}%')
```

## Задача 2

**Оценка доверительного интервала**

Вы исследуете средний рост студентов университета.

Вы измерили рост случайно выбранных студентов и хотите оценить, насколько реальное распределение роста соответствует вашим наблюдениям.

Вы предполагаете, что рост студентов нормально распределен с средним значением 175 см и стандартным отклонением 6 см.

Оцените 95% доверительный интервал для среднего роста студентов и визуализируйте распределение с помощью KDE-графика.

```python
np.random.seed(105)

mean_height = 175
std_dev_height = 6
sample_size = 50

height_sample = np.random.normal(loc=mean_height, scale=std_dev_height, size=sample_size)

sample_mean = np.mean(height_sample)

confidence_level = 0.95
degrees_freedom = sample_size - 1
sample_standard_error = stats.sem(height_sample)
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

print(f'Среднее значение роста выборки: {sample_mean:.2f} см')
print(f'95% доверительный интервал для среднего роста: {confidence_interval[0]:.2f} см - {confidence_interval[1]:.2f} см')

plt.figure(figsize=(12, 6))
sns.kdeplot(height_sample, fill=True)
plt.title('KDE-график распределения роста студентов')
plt.xlabel('Рост (см)')
plt.ylabel('Плотность')
plt.axvline(confidence_interval[0], color='red', linestyle='--', linewidth=2, label=f'Нижняя граница доверительного интервала 95% ({confidence_interval[0]:.2f} см)')
plt.axvline(confidence_interval[1], color='red', linestyle='--', linewidth=2, label=f'Верхняя граница доверительного интервала 95% ({confidence_interval[1]:.2f} см)')
plt.axvline(sample_mean, color='blue', linestyle='-', linewidth=2, label=f'Среднее значение ({sample_mean:.2f} см)')
plt.legend()
plt.show()
```

## Задача 3

**Точечная оценка**

Нужно смоделировать распределение оценок студентов по определенному предмету.

Из предыдущих лет известно, что средняя оценка по этому предмету составляет 75 баллов с стандартным отклонением 10 баллов.

Мы хотим сгенерировать выборку из 100 оценок, чтобы проанализировать, как могут распределяться оценки текущего года.

Посчитайте точечную оценку.

```python
np.random.seed(105)

mean_score = 75
std_dev_score = 10
sample_size = 100

grades = np.random.normal(loc=mean_score, scale=std_dev_score, size=sample_size)

mean_sample = np.mean(grades)

print("Случайная выборка оценок студентов:", grades)
print("Точечная оценка среднего значения выборки:", mean_sample)
```
