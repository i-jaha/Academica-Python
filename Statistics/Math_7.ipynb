## Часть 1 - чистые данные

Библиотека seaborn работает с некоторыми известными датасетами, на которых мы можем опробовать описательный анализ данных.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Мы выбрали датасет mpg, описывающий некоторые из машин бывших в продаже на американском рынке. Довольно интересная информация здесь - характеристика mpg - miles per gallon, сколько миль автомобиль может проехать на галлоне бензина.

```python
data=sns.load_dataset('mpg')
```

```python
data.head()
```

## Задача 1

### 1.1 Посчитайте среднее значение mpg

```python
mpg_mean = data['mpg'].mean()
print("Среднее значение 'mpg':", mpg_mean)
```

### 1.2 Посчитайте медиану mpg

```python
mpg_median = data['mpg'].median()
print("Медиана значений 'mpg':", mpg_median)
```

## Задача 2

### Нарисуйте гистограмму величины мpg, используя seaborn.

```python
sns.histplot(data['mpg'], bins=20, kde=True)
plt.title('Гистограмма значений MPG')
plt.xlabel('MPG')
plt.ylabel('Частота')
plt.show()
```

## Задача 3

### 3.1 Нарисуйте 3 ящика с усами, показывающие распределение mpg в зависимости от региона, в котором машина была произведена

Используйте boxplot, давая две переменные х и y. Если вам нужна будет помощь, посмотрите информацию по ссылке: https://seaborn.pydata.org/examples/grouped_boxplot.html

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', data=data)
plt.title('Распределение MPG в зависимости от региона')
plt.xlabel('Регион')
plt.ylabel('MPG')
plt.show()
```

1. Больше экономичности у японских машин.
2. Большее разнообразие у американских машин.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='cylinders', y='mpg', data=data)
plt.title('Распределение MPG в зависимости от количества цилиндров')
plt.xlabel('Количество цилиндров')
plt.ylabel('MPG')
plt.show()
```

3. Диапазон значений MPG более вариативен у автомобилей с 4 и 5 цилиндрами по сравнению с автомобилями с 3, 6 или 8 цилиндрами. Это может указывать на большое разнообразие моделей и характеристик в этих категориях.

### 4.1 Нарисуйте scatter plot количества лошадинных сил против рабочего объема (displacement). Что вы думаете об их связи между собой?

2. Точки данных на графике образуют примерно линейную тенденцию, что указывает на связь между рабочим объемом и количеством лошадиных сил. Это означает увеличение рабочего объема, что ведет к пропорциональному увеличению мощности.

### 4.2 Нарисуйте scatter plot  рабочего объема против mpg. Что вы думаете об их связи между собой?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='displacement', y='mpg', data=data)
plt.title('Связь между рабочим объемом и экономией топлива (MPG)')
plt.xlabel('Рабочий объем')
plt.ylabel('MPG')
plt.show()
```

### 6.2 Нарисуйте ящики с усами, показывающие распределение цены в зависимости от района, в котором находится квартира/комната. Видите ли вы проблему?

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='neighbourhood', y='price', data=data)
plt.xticks(rotation=90)
plt.title('Распределение цены в зависимости от района')
plt.xlabel('Район')
plt.ylabel('Цена')
plt.show()
```

1. В некоторых районах цены на аренду значительно выше, чем в других. Это указывает на экономическое неравенство и разницу в доступности жилья в зависимости от местоположения.
2. В некоторых районах можно заметить широкий диапазон цен, что говорит о большом разнообразии вариантов жилья: как дорогих, так и более бюджетных вариантов в одном и том же районе.
3. На графике можно увидеть выбросы, которые представляют собой экстремальные значения цен. Это может указывать на наличие люксовых объектов недвижимости по разным причинам.


### 6.3 Нарисуйте ящики с усами, показывающие распределение цены в зависимости от района, в котором находится квартира/комната, используйте следующую команду в boxplot showfliers=False. Что изменилось?

1. В некоторых районах цены на аренду значительно выше, чем в других. Это указывает на экономическое неравенство и разницу в доступности жилья в зависимости от местоположения.
2. В некоторых районах можно заметить широкий диапазон цен, что говорит о большом разнообразии доступных вариантов жилья. Однако это также может указывать на наличие как дорогих, так и более бюджетных вариантов в одном и том же районе.
3. Различия в медианах и межквартильных размахах между районами подчеркивают различия в типичных уровнях арендной платы. Некоторые районы могут быть более престижными или востребованными, что отражается в более высоких медианах цен.

### 6.4 Создайте гистограмму цен для самого распространенного типа комнат в Амстердаме. С какими проблемами вы столкнулись?

```python
most_common_room_type = data['room_type'].value_counts().idxmax()
filtered_data = data[data['room_type'] == most_common_room_type]

plt.figure(figsize=(10, 6))
sns.histplot(filtered_data['price'], bins=20, kde=True)
plt.title(f'Гистограмма цен для самого распространенного типа комнат: {most_common_room_type}')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

1. Гистограмма показывает, что большинство цен сосредоточено в нижней части диапазона до 1000 условных единиц, с резким пиком около низких цен и длинным хвостом, простирающимся к более высоким ценам. Это указывает на то, что большинство квартир или комнат имеют относительно низкие цены, но есть и несколько объектов с очень высокими ценами.
2. Длинный хвост на гистограмме указывает на наличие выбросов, то есть объектов с экстремально высокими ценами. Эти выбросы могут искажать общее представление о распределении цен и затруднять анализ.
3. Гистограмма показывает неравномерное распределение цен, что может указывать на значительные различия в качестве и характеристиках жилья. Это может затруднить потенциальным арендаторам выбор подходящего варианта.
