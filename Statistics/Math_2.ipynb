1. Создайте функцию f(x) = x^6 + 10 * x^3 - 1000

```python
def f(x):
  return x**6 + 10 * x**3 - 1000
```

2. Создайте список x_points (или массив через numpy), в котором у нас лежат числа от -10 до 10

```python
import numpy as np

x_points = np.arange(-10, 11)
x_points
```

3. Используйте x_points как аргумент для созданной функции f(x), и выведите на экран результаты функции для каждого значения x_points, также сохраните результаты в y_points (cписок или массив)

```python
y_points = []
for x in x_points:
  y = f(x)
  print(f"f({x}) = {y}")
  y_points.append(y)
```

4. Покажите на графике x_points & y_points

```python
import matplotlib.pyplot as plt

plt.plot(x_points, y_points)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = x^6 + 10 * x^3 - 1000')
plt.grid(True)
plt.show()
```

5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)

```python
import matplotlib.pyplot as plt
import sympy

x = sympy.Symbol('x')
f_x = x**6 + 10 * x**3 - 1000
f_diff_x = sympy.diff(f_x, x)
print(f"Производная функции f(x): {f_diff_x}")
```

```python
def f_diff(x):
  return 6 * x**5 + 30 * x**2
```

6. Чему равна производная при x = 10

```python
x_val = 10
derivative_at_10 = f_diff(x_val)
print(f"Производная при x = {x_val}: {derivative_at_10}")
```

7. Покажите на графике f(x) & f_diff(x)

```python
x_points_extended = np.linspace(-10, 10, 400) # More points for smoother curve
y_points_extended = [f(x) for x in x_points_extended]
y_diff_points = [f_diff(x) for x in x_points_extended]

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.plot(x_points_extended, y_points_extended, label='f(x)')
plt.plot(x_points_extended, y_diff_points, label='f_diff(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики функций f(x) и f_diff(x)')
plt.legend()
plt.grid(True)
plt.show()
```

8. Найдите уравнение касательной линии в точке x = 9

```python
x_val = 9
y_val = f(x_val)
slope = f_diff(x_val)

# Уравнение касательной: y - y1 = m(x - x1)
# y - y_val = slope * (x - x_val)
# y = slope * (x - x_val) + y_val

print(f"Уравнение касательной в точке x = {x_val}: y = {slope} * (x - {x_val}) + {y_val}")
```

9. Покажите на графике функцию f(x) & касательную линию в точке x = 9

```python
import matplotlib.pyplot as plt
import numpy as np

step = (11 - 7) / (100 - 1)
x_tangent = [7 + i * step for i in range(100)]
x_tangent = np.array(x_tangent)
y_tangent = slope * (x_tangent - x_val) + y_val

plt.figure(figsize=(10, 6))
plt.plot(x_points_extended, y_points_extended, label='f(x)')
plt.plot(x_tangent, y_tangent, label='Tangent at x = 9', color='red') # Plot the tangent line
plt.scatter(x_val, y_val, color='red', zorder=5) # Highlight the point of tangency
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции f(x) и касательной в точке x = 9')
plt.legend()
plt.grid(True)
plt.show()
```
